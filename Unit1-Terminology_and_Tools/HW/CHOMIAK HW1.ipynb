### Background
**Kinematics** is not dynamics, it is just a *description* of how 2D vector fields vary in 2D. 

* Assume LINEAR dependence of u,v on x,y 
    * This involves 4 coefficients $c_1, c_2, c_3, c_4$
* Also, add a constant (not variying with x,y) vector field U,V 

$$ u = U + c_1 x  +c_2 y$$
$$ v = V + c_3 x  +c_4 y$$

## Exercise 1 Answer:
* $ div = \frac{\partial u}{\partial x} + \frac{\partial v}{\partial y}= \frac{\partial u}{\partial x} (U + c_{1}x + c_{2}y) + \frac{\partial v}{\partial y} (V + c_{3}x + c_{4}y) = (0 + c_{1} + 0) + (0 + 0 + c_{4}) = c_{1} + c_{4} $
* $ \zeta_{rel} = \frac{\partial ?}{\partial ?} + \frac{\partial ?}{\partial ?} = \frac{\partial v}{\partial x} - \frac{\partial u}{\partial y}=  \frac{\partial v}{\partial x} (V + c_{3}x + c_{4}y) - \frac{\partial u}{\partial y} (U + c_{1}x + c_{2}y) = (0 + c_3 + 0) - (0 + 0 + c_2) = c_3 - c_2$
* $ def1 = \frac{\partial ?}{\partial ?} + \frac{\partial ?}{\partial ?} = \frac{\partial u}{\partial x} - \frac{\partial v}{\partial y} = \frac{\partial u}{\partial x} (U + c_{1}x + c_{2}y) - \frac{\partial v}{\partial y} (V + c_{3}x + c_{4}y) = (0 + c_{1} +0) - (0 + 0 + c_{4}) = c_1 - c_4$
* $ def2 = \frac{\partial ?}{\partial ?} + \frac{\partial ?}{\partial ?} = \frac{\partial u}{\partial y} + \frac{\partial v}{\partial x} = \frac{\partial u}{\partial y} (U + c_{1}x + c_{2}y) + \frac{\partial v}{\partial x} (V + c_{3}x + c_{4}y) = (0 + 0 + c_2) + (0 + c_3 + 0) = c_2 + c_3$

## Exercise 2: Create and display the basic fields
* First, execute a code cell setting up a coordinate system (x,y) from -10 to 10
* Next, execute the cell defining the basic "pure" div, vort, def1, def2, and U,V fields 
* Finally, display them. The first one has been done for you. You complete the work. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up a coordinate grid centered on 0,0
n = 10
x,y = np.mgrid[-n:n+1, -n:n+1]
```

```python
# Pure divergence of unit magnitude, du/dx + dv/dy
udiv,vdiv = 1*x, 1*y

# Pure vorticity of unit magnitude, du/dy-dv/dx
uvor,vvor = -1*y, 1*x

# Pure def1 of unit magnitude
udef1,vdef1 = 1*x, -1*y

# Pure def2 of unit magnitude
udef2,vdef2 = 1*y, 1*x

# Pure U and V
uU, vU = 1, 0
uV, vV = 0, 1
```

```python
# DEF2

udisp,vdisp = 2*uU + udef2, 2*vU + vdef2
label = 'Pure unit def2 plus U=2'


# Display the vector field as arrows, CENTERED ON the point they apply to
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()
```

## Answer to Exercise 5: 

**When you combine the two deformation components, def1 and def2, you get both divergence and convergence, with little to zero flow at the origin, i.e. when U and V components are both zero. The flow converges as you move towards the origin, then diverges at and away from the origin. This situation is representative of a situation where a low pressure zone could form, the flux of wind leaving the origin (diverging) is greater than than the flux of wind converging at the origin, therefore a low pressure zone would result.**
